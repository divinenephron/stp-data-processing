Scenario 2
====

Instructions
----
The metabolic section are considering offering a disialotransferrin assay instead of a carbohydrate deficient transferrin (CDT) assay to detect heavy alcohol use. Both of these are quantified by the current electrophoresis method, though only CDT is reported. Please compare these measurements using a backup of last week’s results.

Load data
----

```r
# First select this box and press shift-enter to load the data
suppressMessages(library(tidyverse))
data <- read.csv("data/scenario_2.csv")
```

Code
----

```r
# Then select this box and press shift-enter to run it
data %>%
mutate(difference = cdt - sialo_2) %>%
summarise(max = max(difference), min = min(difference), mean = mean(difference))
```

```r
# Really this comparison is best done using a graph, so here's a graph just to show you.
data %>%
ggplot(aes(y = cdt, x = sialo_2)) +
geom_point() +
geom_abline(slope=1, intercept = 0)
```
